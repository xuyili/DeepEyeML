# 代码第一课：代码实现线性回归的梯度下降

````python
import numpy as np
import matplotlib.pyplot as plt
````

```
求$$y=x^2+2x+5$$最小值
先画函数图像
```

````python
x=np.linspace(-100,100,10000) #从-100到100取10000个值
y=x**2+2*x+5
````

````python
plt.plot(x,y) #画图
````

```
初始化x和a，以及迭代次数
```

````python
x=3 #x值
a=0.08 #学习率，步长
m=1000 #迭代次数
````

````python
for i in range(m):
    x = x -a*(2*x+2)
print(x)
````
